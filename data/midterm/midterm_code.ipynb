```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib

df = pd.read_csv("star_stats.csv")
df.head
```

```python
data = df[["temperature", "luminosity"]]
data.head
plt.yscale("log")
plt.xscale("log")
x = df["temperature"]
y = df["luminosity"]


plt.xlabel("temperature")
plt.ylabel("luminosity")
plt.scatter(x,y)

print(df["star_type"].unique())
```

```python
data = df[["temperature", "luminosity"]]
brown_dwarves = df[df["star_type"] == "Brown Dwarf"]
red_dwarves = df[df["star_type"] == "Red Dwarf"]
white_dwarves = df[df["star_type"] == "White Dwarf"]
main_sequences = df[df["star_type"] == "Main Sequence"]
super_giants = df[df["star_type"] == "Super Giant"]
hyper_giants = df[df["star_type"] == "Hyper Giant"]



plt.yscale("log")
plt.xscale("log")
plt.xlabel("temperature")
plt.ylabel("luminosity")



plt.scatter(brown_dwarves["temperature"], brown_dwarves["luminosity"], c="black")
plt.scatter(red_dwarves["temperature"], red_dwarves["luminosity"], c="red")
plt.scatter(white_dwarves["temperature"], white_dwarves["luminosity"], c="grey")
plt.scatter(main_sequences["temperature"], main_sequences["luminosity"], c="blue")
plt.scatter(super_giants["temperature"], super_giants["luminosity"], c="orange")
plt.scatter(hyper_giants["temperature"], hyper_giants["luminosity"], c="purple")





```

```python
sns.kdeplot(df["temperature"])
plt.show()
plt.clf()
sns.kdeplot(df["luminosity"])
plt.show()
plt.clf()
sns.kdeplot(df["radius"])
plt.show()
plt.clf()
sns.kdeplot(df["absolute_magnitude"])
plt.show()
print(df["absolute_magnitude"].var())
```
