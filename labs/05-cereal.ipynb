## Exercise 4: America
Quantify the relationship between sugar and ratings.

Make a plot to visualize this relationship. Superimpose a best fit line (with seaborn) to describe the relationship. It may be helpful to look at the [seaborn regplot documentation]("https://seaborn.pydata.org/generated/seaborn.regplot.html").

```python
# Make the plot for the data visualization and line of best fit here:
sugar_and_ratings = df[["sugars","rating"]]
sns.regplot(data=sugar_and_ratings, x="sugars", y="rating")
```


Calculate a correlation statistic describing the relationship between sugar and ratings  (i.e., r or r squared).

```python
# Calculate the statistic using this cell:
sugar_and_ratings.corr()
print(f"correlation statistic (r-value) = {sugar_and_ratings["sugars"].corr(sugar_and_ratings["rating"])}")

```

Write a statement in plain English interpreting this statistic.

The r-value is a measurement of how closely two features are related (on a scale of -1 to 1), or how closely data trends follow each other. The closer the value is to 1, the more closely the data features are connected in a positive direction (as x goes up, how closely does y tend to go up). The closer the value is to -1, the more closely the data features are connected in a negative direction (as x goes down, how much does y tend to go down.)


## Exercise 5: America Part 2
Make five plots comparing the relationships of carbo, sugars, calories, protein, and fat with rating.

```python
# Write your code to compare the various variables with rating below:
print(f"correlation statistic for carbo = {df["carbo"].corr(df["rating"])}")
print(f"correlation statistic for sugars = {df["sugars"].corr(df["rating"])}")
print(f"correlation statistic for calories = {df["calories"].corr(df["rating"])}")
print(f"correlation statistic for protein = {df["protein"].corr(df["rating"])}")
print(f"correlation statistic for fat = {df["fat"].corr(df["rating"])}")

carbo_plot = sns.lineplot(x=df["carbo"],y=df["rating"])
plt.figure()
sugars_plot = sns.lineplot(x=df["sugars"], y=df["rating"])
plt.figure()
calories_plot = sns.lineplot(x=df["calories"], y=df["rating"])
plt.figure()
protien_plot = sns.lineplot(x=df["protein"], y=df["rating"])
plt.figure()
fat_plot = sns.lineplot(x=df["fat"], y=df["rating"])

```

Sugars and rating have the strongest correlation, as they have the closest r value to 1 or -1 at -0.75.

## Exercise 7: Its Hot and It's Cold
The type column has two values: H='hot' and C='cold'. What is the average rating of each type?

```python
# Write your code to get the average rating by type of cereal below:
hot = df[df["type"] == "H"]
cold = df[df["type"] == "C"]
cold_rating = cold["rating"].mean()
hot_rating = hot["rating"].mean()

print(f"Average hot rating = {hot_rating}")
print(f"Average cold rating = {cold_rating}")
```
